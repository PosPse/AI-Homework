```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_data(label_names: list):
    dataset_path = './dataset/hand-drawn-image-dataset/'
    dataset_dict = {}
    for label_name in label_names:
        dataset_dict[label_name] = np.load(dataset_path + label_name + '/' + label_name + '.npy')
        print(label_name, dataset_dict[label_name].shape)
    return dataset_dict

def show_dataset(dataset_dict: dict, label_names: list):
    for label_name in label_names:
        dataset = dataset_dict[label_name]
        plt.figure(figsize=(22, 2))
        plt.suptitle(label_name)
        random_idx = np.random.randint(0, len(dataset), 10)
        for i, idx in enumerate(random_idx):
            image = dataset[idx]
            plt.subplot(1, 10, i+1)
            plt.imshow(image.reshape(28, 28), cmap='gray')
        plt.show()
```

```python
label_names=['ambulance','apple','bear','bicycle','bird','bus','cat','foot','owl','pig']
dataset_dict = load_data(label_names=label_names)
ambulance_dataset, apple_dataset, bear_dataset, bicycle_dataset, bird_dataset, bus_dataset, cat_dataset, foot_dataset, owl_dataset, pig_dataset = [dataset_dict[label_name] for label_name in label_names]
show_dataset(dataset_dict=dataset_dict, label_names=label_names)
```
